```python
from __future__ import print_function, division
import numpy as np
import scipy as sp

# plotting packages
%matplotlib inline
from matplotlib.pyplot import * # plotting functions
import seaborn as sns           # data visualization package

from analysis_fns import load_emdat, crnplt
```

```python
# load data
select = ["posp","tonic", "phasic"]
data = {}

for s in select:
    res, mp = load_emdat(s,EOcut=0.2,CORcut=0.2,root="/home/tyler/dstrf_results/modeldata/")
    data[s] = {"results": res,
               "map": mp}
```

```python
clr = sns.cubehelix_palette(3,reverse=False,light=0.15,dark=0.7,rot=1,start=90)
```

```python
# make corner plot
sns.set_style("ticks")
sns.set_palette(clr)

figure(figsize=(10,10))

matlabs =  [r'$\omega$',r'$\alpha_1$',r'$\alpha_2$']

mrange=[[  0  , 20  ],
        [-25.1, 25.1],
        [ -1.1,  1.1]],

HIST = False

for s in select:
    crnplt(data[s]["map"],s,mrange=mrange,labs=matlabs,hist=HIST)

suptitle("Population Parameter Distribution",y=1.04)
```
