```python
from __future__ import print_function, division
import numpy as np
import scipy as sp

# plotting packages
%matplotlib inline
from matplotlib.pyplot import * # plotting functions
import seaborn as sns           # data visualization package

from analysis_fns import load_emdat, crnplt
```

```python
# load data
posp_results, posp_maxlik = load_emdat("posp")
tonic_results, tonic_maxlik = load_emdat("tonic")
phasic_results, phasic_maxlik = load_emdat("phasic")
```

```python
# make corner plot
sns.set_style("ticks")
sns.set_palette("cubehelix",3)

figure(figsize=(10,10))

matlabs =  [r'$\omega$',r'$\alpha_1$',r'$\alpha_2$']

mrange=[[  0  , 20  ],
        [-25.1, 25.1],
        [ -1.1,  1.1]],


HIST = False

crnplt(posp_maxlik,"posp",mrange=mrange,labs=matlabs,hist=HIST)
crnplt(tonic_maxlik,"tonic",mrange=mrange,labs=matlabs,hist=HIST)
crnplt(phasic_maxlik,"phasic",mrange=mrange,labs=matlabs,hist=HIST)

suptitle("Population Parameter Distribution",y=1.04)
```
